**Load data from a database - MySQL**

```python
import pandas as pd
import mysql.connector

# Connect to MySQL database
cnx = mysql.connector.connect(
  host="<your_host>",
  user="<your_username>",
  password="<your_password>",
  database="<your_database>"
)

# Query data from database
df = pd.read_sql_query("SELECT * FROM <your_table>", cnx)

# Close database connection
cnx.close()

# Display the first 5 rows of the dataset
print(df.head())
```

**Load data from a database - MSSQL**

```python
import pyodbc
import pandas as pd

# Define the database connection parameters
server = '<your_server>'
database = '<your_database>'
username = '<your_username>'
password = '<your_password>'
driver= '{ODBC Driver 17 for SQL Server}'

# Define the connection string
cnxn = pyodbc.connect('DRIVER='+driver+';SERVER='+server+';DATABASE='+database+';UID='+username+';PWD='+ password)

# Query data from database
query = 'SELECT * FROM <your_table>'
df = pd.read_sql(query, cnxn)

# Close database connection
cnxn.close()

# Display the first 5 rows of the dataset
print(df.head())
```

**Load data  set from MongoDB**

```python
import pymongo
import pandas as pd

# Define the connection parameters
uri = "<your_mongodb_uri>"
client = pymongo.MongoClient(uri)
db = client.get_database("<your_database>")
collection = db.get_collection("<your_collection>")

# Query data from MongoDB collection
cursor = collection.find({})
df = pd.DataFrame(list(cursor))

# Display the first 5 rows of the dataset
print(df.head())

# Close MongoDB connection
client.close()
```
