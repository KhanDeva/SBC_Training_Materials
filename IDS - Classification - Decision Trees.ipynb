# Decision Trees for Classification

It is a tree-like model where each internal node represents a decision based on one of the input features, and each leaf node represents a class label or a numeric value.
The goal of a Decision Tree algorithm is to split the input space into regions, where each region corresponds to a different class label or a numeric value.
The algorithm selects the input feature that provides the best split of the data based on some criterion, such as information gain or Gini impurity using attribute selection measure(ASM). 
The root node is selected based on the attribute selection measure (ASM) and this is repeated until there is a leaf node. 
ASM is a technique used in data mining processes for data reduction. The main ASM techniques are Gini Index, Gain ratio and Information Gain (ID3).


```python
#load data set
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
#split the data set to training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#implementation of the decision tree
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Fit a decision tree classifier to the data
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

```python
#make a prediction
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
