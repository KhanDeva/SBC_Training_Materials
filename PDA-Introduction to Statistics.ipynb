```python
#Descriptive Measures
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
                 names=["sepal_length", "sepal_width", "petal_length", "petal_width", "class"])

# Display the first few rows of the dataset
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Compute summary statistics for each column
print(df.describe())

# Compute the correlation matrix
corr_matrix = df.corr()

# Visualize the correlation matrix as a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()

# Create a scatter plot of two columns
sns.scatterplot(x="sepal_length", y="petal_length", hue="class", data=df)
plt.show()

# Create a boxplot of a column
sns.boxplot(y="petal_width", x="class", data=df)
plt.show()

# Create a violinplot of two columns
sns.violinplot(y="sepal_width", x="class", data=df)
plt.show()
```

```python
#Measures of Skewness
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',
                 header=None, names=['class', 'alc', 'malic', 'ash', 'alc_ash', 'mag', 'tot_phen', 
                                     'flav', 'nonflav_phen', 'proant', 'color_int', 'hue', 'od280', 'proline'])

# Calculate the skewness of each column
skewness = df.skew()

# Print the skewness values
print(skewness)

# Visualize the distribution of each column using histograms
df.hist(figsize=(10, 10))
plt.show()

# Visualize the skewness of each column using boxplots
sns.boxplot(data=df)
plt.show()
```

```python
#Kurtosis 
#Kurtosis is a measure of the "peakedness" or "flatness" of a probability distribution. 
#It is a statistical measure that describes the degree to which the tails of a distribution differ from the tails of a normal distribution. 
#Kurtosis is often used to assess whether a distribution is normal or not.
#A distribution with high kurtosis has more of its values concentrated around the mean, and has heavier tails than a normal distribution. 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',
                 header=None, names=['class', 'alc', 'malic', 'ash', 'alc_ash', 'mag', 'tot_phen', 
                                     'flav', 'nonflav_phen', 'proant', 'color_int', 'hue', 'od280', 'proline'])

# Calculate the kurtosis of the 'alc' column
kurtosis = df['alc'].kurtosis()

# Print the kurtosis value
print('Kurtosis of the \'alc\' column:', kurtosis)

# Visualize the distribution of the 'alc' column using a histogram
df['alc'].hist()
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.title('Distribution of Alcohol Content')
plt.show()
```

```python
#Kurtosis - a distribution with low kurtosis has fewer values around the mean, and has lighter tails than a normal distribution. Such a distribution is called a platykurtic distribution.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',
                 header=None, names=['class', 'alc', 'malic', 'ash', 'alc_ash', 'mag', 'tot_phen', 
                                     'flav', 'nonflav_phen', 'proant', 'color_int', 'hue', 'od280', 'proline'])

# Calculate the kurtosis of the 'proline' column
kurtosis = df['proline'].kurtosis()

# Print the kurtosis value
print('Kurtosis of the \'proline\' column:', kurtosis)

# Visualize the distribution of the 'proline' column using a histogram
df['proline'].hist()
plt.xlabel('Proline')
plt.ylabel('Frequency')
plt.title('Distribution of Proline')
plt.show()
```
